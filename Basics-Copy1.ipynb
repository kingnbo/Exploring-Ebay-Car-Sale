```python
import pandas as pd
import numpy as np
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names = cols)
pd.options.display.max_columns = 99
```

```python
cars = cars.replace('?',np.nan)
```

```python
numerical_col = ['normalized-losses','wheel-base','length','width','height','curb-weight','bore','stroke','compression-rate','horsepower','peak-rpm','city-mpg','highway-mpg','price']
cars_numerical = cars[numerical_col]
```

```python
cars_numerical.info()
```

```python
cars_numerical = cars_numerical.astype(float)
cars_numerical.info()
```

```python
cars_numerical.isnull().sum()
```

```python
cars_numerical = cars_numerical.dropna(subset =['price'])
cars_numerical.info()
```

```python
cars_numerical.isnull().sum()
```

```python
cars_numerical = cars_numerical.fillna(cars_numerical.mean())
cars_numerical.isnull().sum()
```

```python
price_col = cars_numerical['price']
```

```python
cars_numerical = (np.abs(cars_numerical-cars_numerical.mean()))/(cars_numerical.max()-cars_numerical.min())
cars_numerical['price'] = price_col
```

```python
cars_numerical
```

```python
%matplotlib inline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

k_list = [5] #k-value

def knn_train_test(training_col,target_col,df):
    
    K_RMSE_MAP = {} # dictionnary to store RMSE values for each k value
    
    for k in k_list:# this for loop is necessary since k is in a list 
        
        # initialize the nearest neighbor object.
        knn = KNeighborsRegressor(n_neighbors=k)
        
        np.random.seed(1)#--> this makes sure we get the same random order everytime so our reults don't change


        mixed_index = np.random.permutation(df.index)#-->mix the index of our dataframe(i.e. mix the row numbers)
        random_df = df.reindex(mixed_index)# reindex the data (i.e. mix the rowsand applys to the dataframe)

        last_train_row = int(len(random_df)/2)

        train = random_df.iloc[0:last_train_row]
        test = random_df.iloc[last_train_row:]


        knn.fit(train[training_col],train[target_col])
        predictions = knn.predict(test[training_col])
        mse = mean_squared_error(test[target_col],predictions)
        K_RMSE_MAP[k]= mse**(1/2)
    
    return K_RMSE_MAP
train_cols = cars_numerical.columns.drop('price')
rmses = {}


for feature in train_cols:
        rmses[feature] =  knn_train_test([feature],['price'],cars_numerical)
print(rmses)
rmse_values ={}

for title,rmse_dict in rmses.items():
    rmse_values[title] = list(rmse_dict.values())
    
rmse_df = pd.DataFrame.from_dict(rmse_values)

def get_n_best_features(n,rmse):
    
    return rmse.iloc[0].nsmallest(n)
        
print('The best two features are:')    
best_two_features = list(get_n_best_features(2,rmse_df).index)
print(best_two_features)

print('The best three features are:') 
best_three_features = list(get_n_best_features(3,rmse_df).index)
print(best_three_features)

print('The best four features are:') 
best_four_features = list(get_n_best_features(4,rmse_df).index)
print(best_four_features)

print('The best five features are:') 
best_five_features = list(get_n_best_features(5,rmse_df).index)
print(best_five_features)
```

```python
print('The best two features give the RMSE:')
print(knn_train_test(best_two_features,['price'],cars_numerical))

print('The best three features give the RMSE:')
print(knn_train_test(best_three_features,['price'],cars_numerical))

print('The best four features give the RMSE:') 
print(knn_train_test(best_four_features,['price'],cars_numerical))

print('The best five features give the RMSE:') 
print(knn_train_test(best_five_features,['price'],cars_numerical))
```

The three best models are the models that use 3, 2 and 4 features in that order.
Let's change the hyperparameter K from 1-25 to find the minimum RMSE
